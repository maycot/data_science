```python
% matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from os import walk
```

# MEAN / BATCH 100

```python
path = './comp1/'
filename_list = []
for filename in next(walk(path))[2]:
    filename_list.append(filename)
filename_list.remove('.DS_Store')
```

```python
filename_list
```

```python
df_all = pd.DataFrame()
for filename in filename_list:
    df = pd.read_csv(path + filename).set_index("Unnamed: 0")
    df.columns = [filename[-8:-4]]
    df_all = pd.concat([df_all, df], axis=1)
```

```python
batch = ['ch_0', '_100', '_200', '_300', '_400', '_500', '_600']
figsize = (20,15)
ncol = 1
nrow = len(batch)
fig, axes = plt.subplots(nrow, ncol, figsize=figsize)
plt.tight_layout()
k=0
for j in batch:
    if k%2==0:
        c='seagreen'
    else:
        c='red'
    df_all.plot(y = j, ax=axes[k], kind='bar', xticks=[], ylim = [-0.1, 0.2], color=c)
    k=k+1
```
