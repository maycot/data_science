```python
df_all = pd.DataFrame()
for filename in filename_list:
    df = pd.read_csv(path2 + filename).set_index("Unnamed: 0")
    df.columns = [filename[-8:-4]]
    df_all = pd.concat([df_all, df], axis=1)
```

```python
l1.remove('h_0')
```

```python
l2 = ['_' + str(el) for el in range(100,650,10)]
```

```python
df_all.columns
```

```python
batch = ['ch_0'] + l1 + l2
figsize = (20,60)
ncol = 1
nrow = len(batch)
fig, axes = plt.subplots(nrow, ncol, figsize=figsize)
plt.tight_layout()
k=0
for j in batch:
    if k%2==0:
        c='seagreen'
    else:
        c='red'
    df_all.plot(y = j, ax=axes[k], kind='bar', xticks=[], ylim = [-10, 10], color=c)
    k=k+1
```
