**Test Anova sur variable quantitative nombre d'amis avant 30 jours**

```python
categorical_features = df.columns.drop(['nb_ami', 'nb_ami_avant_30', 'nb_ami_avant_10', 'is_ami_avant_30',
                                        'salaire_annuel_normalise', 'dd_datecreationreport', 'kc_offre',
                                        'dc_intituleoffre', 'dc_descriptifoffre', 'dc_descriptifentreprise'])
```

```python
categorical_features_to_test = categorical_features.drop(['dc_codepostallieutravail', 'dc_rome_id', 'dc_appelationrome_id',
                                                         'dc_specificite1', 'dc_specificite2', 'dc_specificite3'])
```

```python
def anova_test(data: pd.DataFrame, y: str, x: str):
    
    print(f'\n\n**** Fisher test {x} on {y} ****\n')
    # boxplot if less than 150 classes
    if len(set(df[x])) < 150:
        sns.set()
        plt.figure(figsize=(20, 4))
        ax = sns.boxplot(x=x, y=y, data=data)
        plt.xlabel(x)
        plt.ylabel(y)
        plt.show()
        
    anova = smf.ols(f'{y}~{x}', data=data).fit()
    #print(anova.summary())
    print(sm.stats.anova_lm(anova, typ=2))
```

```python
data = df[df.nb_ami_avant_30 < 50]
y = "nb_ami_avant_30"

for x in categorical_features_to_test:
    anova_test(data, y, x)
```

**Test Cramer sur variable qualitative "avec ami avant 30 jours"**

```python
def cramers_corrected_stat(confusion_matrix):
    """ calculate Cramers V statistic for categorial-categorial association.
        uses correction from Bergsma and Wicher, 
        Journal of the Korean Statistical Society 42 (2013): 323-328
    """
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))    
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min( (kcorr-1), (rcorr-1)))
```

```python
categorical_features_to_test = list(categorical_features_to_test) + ['outre_mer', 'dc_specificite1', 'dc_specificite2', 'dc_specificite3']
```

```python
cramer = pd.DataFrame(index=categorical_features_to_test, columns=["cramer_v"])

# convert to categories type
for col in categorical_features_to_test:
    df[col] = df[col].astype('category')
    cramer.loc[col, "cramer_v"] = cramers_corrected_stat(pd.crosstab(df.is_ami_avant_30, df[col]))
```

```python
cramer.sort_values(by="cramer_v", ascending=False).plot(kind="bar", figsize=(10, 5))
plt.title('Cramer test on AMI_AVANT_30_JOURS')
plt.show()
```

#### Conclusion Test de Cramer

predominent categorical features on target ami_avant_30:
