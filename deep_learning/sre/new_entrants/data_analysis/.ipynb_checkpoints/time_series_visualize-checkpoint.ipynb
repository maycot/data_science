```python
%matplotlib inline
    
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

import sys
sys.path.append('..')

from models.ml.utils import load_csv_to_df

import warnings
warnings.filterwarnings('ignore')

path = "../models/data"
filename_2016 = "train_set.csv"
filename_2017 = "test_set.csv"
filename_all = "dataset.csv"
```

## Visualize Time Series

    Compare 2016 and 2017 data

    Goal : define best strategy to split train/test set

    Find outliers

```python
df_all = load_csv_to_df(path, filename_all)
```

```python
df_all = df_all[df_all.rome.notnull()]
df_all["dep"] = df_all.depcom.apply(lambda x: str(x)[:2])
df_all = df_all[df_all.dep != "na"]
df_all["domaine_pro"] = df_all.rome.apply(lambda x: x[:3])
df_all["datins"] = pd.to_datetime(df_all["datins"])
df_all['month'] = df_all.datins.apply(lambda x: x.month)
df_all['year'] = df_all.datins.apply(lambda x: x.year)
```

```python
df_all = df_all[["ident", "rome", "domaine_pro", "datins", "month", "year", "dep", "ict1"]]
df_all = df_all[df_all.dep != "98"]
```

### Count classes (dep - rome / dep - domaine_pro)

```python
rome_count_2016 = len(set(df_all[df_all.year == 2016].rome))
dep_count_2016 = len(set(df_all[df_all.year == 2016].dep))
print("2016 : Nber of rome : {} / nber of dep : {}".format(rome_count_2016, dep_count_2016))
```

```python
rome_count_2017 = len(set(df_all[df_all.year == 2017].rome))
dep_count_2017 = len(set(df_all[df_all.year == 2017].dep))
print("2017 : Nber of rome : {} / nber of dep : {}".format(rome_count_2017, dep_count_2017))
```

```python
# Number of classes : rome * dep
c2016 = df_all[df_all.year == 2016].reset_index().groupby(["rome", "dep"])["ident"].count().count()
c2017 = df_all[df_all.year == 2017].reset_index().groupby(["rome", "dep"])["ident"].count().count()
print("2016 : Nber of classes rome * dep : {} / 2017 : Nber of classes rome * dep : {}".format(c2016, c2017))
```

```python
# Number of classes : dep * domaine_pro
c2016 = df_all[df_all.year == 2016].reset_index().groupby(["dep", "domaine_pro"])["ident"].count().count()
c2017 = df_all[df_all.year == 2017].reset_index().groupby(["dep", "domaine_pro"])["ident"].count().count()
print("2016 : Nber of classes domainepro * dep : {} / 2017 : Nber of classes domainepro * dep : {}".format(c2016, c2017))
```

### Analyze time series

```python
df_all.set_index("datins", inplace=True)
df_all.ict1 = df_all.ict1.astype("int")
ict1_by_month = df_all.resample("M").sum()
```

```python
ict1_by_month.ict1.plot(title="Global ic", figsize=(10,3))
plt.show()
```

```python
ict1_by_month_2016 = df_all[df_all.year == 2016].ict1.resample("M").sum()
ict1_by_month_2017 = df_all[df_all.year == 2017].ict1.resample("M").sum()
```

```python
ax = ict1_by_month_2016.plot(use_index = False)
ict1_by_month_2017.plot(ax=ax, use_index = False, title="Global ic : Compare 2016 and 2017", figsize=(10,5))
plt.show()
```

### ict1 by dep

```python
dep_grouped = df_all.groupby(["dep"])
dep_ict1 = dep_grouped.ict1.resample("M").sum()
```

```python
dep_pivot = pd.pivot_table(pd.DataFrame(dep_ict1), index="datins", columns="dep")
```

```python
ax = dep_pivot.plot(figsize=(25, 10), legend=False)
#ax.legend(loc='upper right', ncol=5, fontsize=12)
ax.set_title("IC BY DEP", fontsize=14)
plt.show()
```

```python
dep_pivot.max()[dep_pivot.max() == dep_pivot.max().max()]
```

```python
dep_pivot.min()[dep_pivot.min() == dep_pivot.min().min()]
```

```python
ax = dep_pivot.iloc[:, 75].plot(use_index = False)
dep_pivot.iloc[:, 44].plot(ax=ax, use_index = False, title="Global ict : Compare dep 75 and 44", figsize=(7,3))
plt.show()
```

### ic by dep and domaine pro

```python
grouped = df_all.groupby(["dep", "domaine_pro"])
grouped_ict1 = grouped.ict1.resample("M").sum()
```

```python
grouped_pivot = pd.pivot_table(pd.DataFrame(grouped_ict1), index="datins", columns=["dep", "domaine_pro"])
```

```python
grouped_pivot.shape
```

```python
res75 = grouped_pivot.iloc[:, grouped_pivot.columns.get_level_values(1)=="75"]
ax = res75.plot(figsize=(25, 7), legend=False, ylim=[0, 750])
ax.set_title(" DEP 75 : ICT BY DOMAINE PRO", fontsize=14)
plt.show()
```
